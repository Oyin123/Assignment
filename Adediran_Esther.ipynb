```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report, precision_score

import optuna
```

```python
data = pd.read_csv("heart_2020_cleaned.csv")
```

```python
labels_to_encode = data.select_dtypes(include="object").columns

encoder = LabelEncoder()
data[labels_to_encode] = data[labels_to_encode].apply(encoder.fit_transform)

data.head()
```

```python
labels_to_scale = data.select_dtypes(include="float").columns

scaler = StandardScaler()
data[labels_to_scale] = scaler.fit_transform(data[labels_to_scale])

data.head()
```

```python
data.HeartDisease.value_counts()
```

```python
data = pd.concat([data.query('HeartDisease == 1'),
                  data.query('HeartDisease == 0').sample(frac=0.1)],
                ignore_index=True)

data.HeartDisease.value_counts()
```

```python
X = data.drop("HeartDisease",1)
y = data["HeartDisease"]
X_train, X_test, y_train, y_test = train_test_split(
    X,y,random_state=43, stratify=y,test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_preds = model.predict(X_test)
```

```python
print(classification_report(y_test,y_preds))
```

```python
def objective(trial):
    solver = trial.suggest_categorical("solver", ["newton-cg", "lbfgs", "liblinear", "sag", "saga"])
    penalty = trial.suggest_categorical("penalty", ["None", "l1", "l2", "elasticnet"])
    c = trial.suggest_float("c", 0.01, 100, log=True)
    clf = LogisticRegression(solver=solver)
    clf.fit(X_train, y_train)
    return precision_score(y_test, clf.predict(X_test))
    
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)

trial = study.best_trial
print('Precision score: {}'.format(trial.value))

print("Best hyperparameters: {}".format(trial.params))
```
